# Pattern : Prefix Sums
Use when you need repeated range sum queries.

## What is a Prefix Sum?

A **prefix sum** (also called cumulative sum) is an array where each element represents the sum of all elements from the start up to that index.

### Example

```
Original Array:     [2,  3,  5,  7]
                     0   1   2   3

Prefix Sum Array:   [2,  5, 10, 17]
                     0   1   2   3

Explanation:
prefix[0] = 2           (just 2)
prefix[1] = 2 + 3 = 5   (sum of 2,3)
prefix[2] = 2+3+5 = 10  (sum of 2,3,5)
prefix[3] = 2+3+5+7 = 17 (sum of all)
```

## Why Use Prefix Sums?

### Problem Without Prefix Sum

Find sum of elements between index i and j (inclusive) for MANY queries.

```python
# Naive approach: O(n) per query
def range_sum_naive(arr, i, j):
    total = 0
    for k in range(i, j + 1):
        total += arr[k]
    return total

# If you have 1000 queries on array of 10,000 elements:
# Time = 1000 * 10,000 = 10,000,000 operations (SLOW!)
```

### Solution With Prefix Sum

```python
# O(n) preprocessing, O(1) per query
def range_sum_fast(prefix, i, j):
    if i == 0:
        return prefix[j]
    else:
        return prefix[j] - prefix[i - 1]

# Same 1000 queries: 10,000 + 1000 = 11,000 operations (FAST!)
```

## Building a Prefix Sum Array

### Method 1: Create New Array (Extra Space)

```python
def build_prefix_sum(arr):
    """Build new prefix sum array - O(n) time, O(n) space"""
    n = len(arr)
    prefix = [0] * n
    
    prefix[0] = arr[0]
    for i in range(1, n):
        prefix[i] = prefix[i - 1] + arr[i]  # Add current element to previous sum
    
    return prefix

arr = [2, 3, 5, 7]
prefix = build_prefix_sum(arr)
print(f"Array:  {arr}")
print(f"Prefix: {prefix}")
```

### Method 2: Modify Original Array (In-place)

```python
def build_prefix_sum_inplace(arr):
    """Modify array to become prefix sum - O(n) time, O(1) space"""
    for i in range(1, len(arr)):
        arr[i] += arr[i - 1]  # Each element becomes cumulative
    return arr

arr = [2, 3, 5, 7]
build_prefix_sum_inplace(arr)
print(f"After in-place: {arr}")  # [2, 5, 10, 17]
```

## Using Prefix Sum for Range Queries

### The Magic Formula

To find sum of elements from index i to j (inclusive):

```
sum(i to j) = prefix[j] - prefix[i-1]
```


### Example

```
Array:        [2,  3,  5,  7]
Prefix:       [2,  5, 10, 17]
Index:         0   1   2   3

Query: Sum from index 1 to 3?
prefix[3] - prefix[0] = 17 - 2 = 15
(which is 3 + 5 + 7 = 15) ✓

Query: Sum from index 2 to 3?
prefix[3] - prefix[1] = 17 - 5 = 12
(which is 5 + 7 = 12) ✓
```

## When to Use Prefix Sums

### Use Prefix Sum When:

1. **Multiple range sum queries** - O(n) preprocessing pays off quickly
2. **Subarray sum equals k** - Find subarrays with target sum
3. **Product of array except self** - Can use prefix/suffix products
4. **Continuous subarray sum** - Check divisibility or specific sums
5. **2D array problems** - 2D prefix sums for rectangle sum queries

### Don't Use When:

1. **Single query** - Just iterate once (O(n) is O(n) anyway!)
2. **Array changes** - Would need to rebuild prefix array each time
3. **Space is critical** - O(n) extra space not allowed

````python
def build_prefix_sum(arr):
    """Build prefix sum array"""
    n = len(arr)
    prefix = [0] * n
    
    prefix[0] = arr[0]
    for i in range(1, n):
        prefix[i] = prefix[i - 1] + arr[i]
    
    return prefix

# Example
original = [2, 3, 5, 7]
prefix = build_prefix_sum(original)

print(f"Original Array:  {original}")
print(f"Prefix Sum:      {prefix}")
print()
````

# Find sum of elements between indices i and j

````python
print(" Range Sum Query Using Prefix")

def range_sum(prefix, i, j):
    """
    Find sum of elements from index i to j (inclusive)
    
    Time: O(1)
    Space: O(1)
    """
    if i == 0:
        return prefix[j]
    else:
        return prefix[j] - prefix[i - 1]

print(f"Sum of elements from index 1 to 3: {range_sum(prefix, 1, 3)}")
````

# Find count of subarrays with sum equal to k

````python
print("Subarray Sum Equals K")

def subarray_sum_equals_k(arr, k):
    """
    Approach: Use prefix sum + hashmap
    - Store count of each prefix sum seen
    - If (current_prefix - k) exists, we found subarrays
    
    Time: O(n)
    Space: O(n)
    """
    count = 0
    current_sum = 0
    sum_count = {0: 1}  # Initialize with 0 having count 1
    
    for num in arr:
        current_sum += num
        
        # If (current_sum - k) exists, we found subarrays
        if current_sum - k in sum_count:
            count += sum_count[current_sum - k]
        
        # Add current sum to map
        sum_count[current_sum] = sum_count.get(current_sum, 0) + 1
    
    return count

# Test
arr = [1, 2, 3, 1, 1, 1]
k = 3

print(f"Array: {arr}")
print(f"Target Sum: {k}")
print()

result = subarray_sum_equals_k(arr, k)
print(f"Count of subarrays with sum = {k}: {result}")
print()

# Show the subarrays
print("The subarrays are:")
n = len(arr)
count = 0
for i in range(n):
    for j in range(i, n):
        subarray_sum = sum(arr[i:j+1])
        if subarray_sum == k:
            count += 1
            print(f"  {count}. arr[{i}:{j+1}] = {arr[i:j+1]} → Sum = {subarray_sum}")

print()
````

#     For each element, find product of all other elements

````python

print("Product of Array Except Self")


def product_except_self(arr):
    """    
    Approach: Use prefix/suffix products
    result[i] = (product of all left) * (product of all right)
    
    Time: O(n)
    Space: O(n) for output (not counting output)
    """
    n = len(arr)
    result = [1] * n
    
    # Build prefix products (product of all elements to the left)
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= arr[i]
    
    # Build suffix products (product of all elements to the right)
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= arr[i]
    
    return result

# Test
arr = [1, 2, 3, 4]
result = product_except_self(arr)

print(f"Array:  {arr}")
print(f"Result: {result}")
print()

print("Explanation:")
for i in range(len(arr)):
    others = arr[:i] + arr[i+1:]
    product = 1
    for x in others:
        product *= x
    print(f"result[{i}] = product of {others} = {result[i]}")

print()
````
