# Pattern 1: Neighbor Sum / Array Transformation

Problem: Given array a , return array b where each b[i] = sum of a[i] and its left/right
neighbors.

````python
def neighbours(a):
    n = len(a)
    b = [0]*len(a)
    for i in range(n):
        b[i] = a[i]  
        if i > 0:
            b[i] += a[i - 1]     # left neighbour
        if i < n - 1:
            b[i] += a[i + 1]    # right neighbour
    return b
    
print(neighbours([2, 3, 4, 5]))
````

## How Extreme Cases Are Handled - Step by Step

### The Key Insight

The conditions `if i > 0:` and `if i < n - 1:` **automatically skip invalid neighbors**. They don't need special handling!

### Example: [10, 20, 30]

```
Array:  [10,  20,  30]
Index:   0    1    2
```

### Iteration by Iteration

#### i = 0 (FIRST element - extreme left)

```python
b[0] = a[0] = 10          # Start with value itself

if i > 0:                 # if 0 > 0: FALSE
    # This block SKIPPED - no left neighbor exists!
    
if i < n - 1:             # if 0 < 2: TRUE
    b[0] += a[0 + 1]      # Add right neighbor
    b[0] = 10 + 20 = 30

Result: b[0] = 30
(Only has right neighbor, no left neighbor - PERFECT!)
```

#### i = 1 (MIDDLE element)

```python
b[1] = a[1] = 20          # Start with value itself

if i > 0:                 # if 1 > 0: TRUE
    b[1] += a[1 - 1]      # Add left neighbor
    b[1] = 20 + 10 = 30

if i < n - 1:             # if 1 < 2: TRUE
    b[1] += a[1 + 1]      # Add right neighbor
    b[1] = 30 + 30 = 60

Result: b[1] = 60
(Has both neighbors - PERFECT!)
```

#### i = 2 (LAST element - extreme right)

```python
b[2] = a[2] = 30          # Start with value itself

if i > 0:                 # if 2 > 0: TRUE
    b[2] += a[2 - 1]      # Add left neighbor
    b[2] = 30 + 20 = 50

if i < n - 1:             # if 2 < 2: FALSE
    # This block SKIPPED - no right neighbor exists!

Result: b[2] = 50
(Only has left neighbor, no right neighbor - PERFECT!)
```

### Final Result

```
Input:  [10, 20, 30]
Output: [30, 60, 50]

Breakdown:
- b[0] = 30 = 10 (itself) + 20 (right)
- b[1] = 60 = 10 (left) + 20 (itself) + 30 (right)
- b[2] = 50 = 20 (left) + 30 (itself)
```
This is called **"defensive programming"** - prevent bad things BEFORE they happen!

# Pattern 2: Two Pointers
Used for problems involving pairs, sorted arrays, or palindromes.

## Check if a string is a palindrome


````python
def palindrome(a):
    left ,right = 0, len(a) - 1
    while left < right:
        if a[left] != a[right]:
            return False
        left += 1
        right -= 1
    return True
print(palindrome("helleh"))
````

## Find two numbers that sum to target (sorted array)

````python
def twosumsorted(a,target):
    left, right = 0, len(a) - 1
    while left < right:
        if a[left] + a[right] == target:
            return a[left], a[right]
        if a[left] + a[right] < target:
            left += 1
        else:
            right -= 1
            
print(twosumsorted([2,3,5,6,7,8,9], 10))
````
