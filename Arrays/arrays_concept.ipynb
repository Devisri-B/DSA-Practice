## Section 1: Array Fundamentals

### What is an Array?

An **array** is a contiguous block of memory that stores elements of the same data type at fixed-size locations. Each element is accessed via an index (0-based in most languages).

### Key Characteristics

- **Contiguous memory**: Elements stored sequentially in memory
- **Fixed size**: Size determined at creation (in some languages like Java/C++)
- **Zero-indexed**: First element at index 0
- **Direct access**: Can access any element in O(1) time
- **Type homogeneous**: All elements same type (in typed languages)

### Array vs List (Python)

```python
# Python lists are dynamic arrays
arr = [1, 2, 3, 4, 5]  # Flexible size, grows automatically

# But they're built on top of array concept
# Internally: contiguous memory that resizes when needed
```

### Time Complexity Summary

| Operation | Time | Why? |
|-----------|------|-----|
| Access by index | O(1) | Direct memory mapping |
| Search (unsorted) | O(n) | Must check each element |
| Search (sorted) | O(log n) | Binary search |
| Insert at end | O(1) | If space available |
| Insert at start | O(n) | Must shift all elements |
| Insert in middle | O(n) | Must shift elements after |
| Delete from end | O(1) | Just remove |
| Delete from start | O(n) | Must shift elements |
| Delete from middle | O(n) | Must shift remaining elements |

### Space Complexity

- **Array of n elements**: O(n)
- **No extra space operations**: O(1) auxiliary space

## Section 2: Common Array Patterns for Interviews

### Pattern 1: Two Pointers

**When to use**: Find pairs, reverse, partition, detect cycles

```
Approach: Use two pointers moving from start/end toward center
Time: O(n)
Space: O(1)
```

**Example problems**:
- Valid palindrome
- Two sum (sorted array)
- Container with most water
- Merge sorted arrays

### Pattern 2: Sliding Window

**When to use**: Find subarray/substring with specific property

```
Approach: Maintain window of elements, expand/contract based on condition
Time: O(n)
Space: O(1) or O(k) depending on problem
```

**Example problems**:
- Longest substring without repeating characters
- Maximum sum of subarray of size k
- Minimum window substring
- Longest subarray with sum = k

### Pattern 3: Binary Search

**When to use**: Array is sorted, find element or boundary

```
Approach: Divide search space in half each time
Time: O(log n)
Space: O(1)
```

**Example problems**:
- Search in sorted array
- Find first/last position of element
- Search in rotated sorted array
- Find peak element

### Pattern 4: Prefix Sum / Cumulative Sum

**When to use**: Range sum queries, subarray problems

```
Approach: Build prefix array, use for O(1) range queries
Time: O(n) preprocessing, O(1) query
Space: O(n)
```

**Example problems**:
- Range sum query
- Subarray sum equals k
- Continuous subarray sum
- Product of array except self

### Pattern 5: Sorting / Partitioning

**When to use**: Reorder elements by some criteria

```
Approach: Use sort or partition algorithm
Time: O(n log n) for sort, O(n) for partition
Space: O(1) or O(n) depending on stability
```

**Example problems**:
- Sort colors (0s, 1s, 2s)
- Wiggle sort
- Move zeros to end
- Separate odd and even

### Pattern 6: DFS / Backtracking

**When to use**: Generate combinations, permutations, subsets

```
Approach: Recursive exploration with backtracking
Time: O(2^n) for subsets, O(n!) for permutations
Space: O(n) recursion depth
```

**Example problems**:
- Permutations
- Combinations
- Subsets
- Word search in 2D array

### Pattern 7: HashMap / HashSet

**When to use**: Need fast lookup, count occurrences

```
Approach: Store elements in hash table for O(1) lookup
Time: O(n)
Space: O(n)
```

**Example problems**:
- Two sum
- Duplicate detection
- Anagrams
- Group anagrams

````python
print("=== Common Array Techniques ===\n")

# Pattern 1: Two Pointers
print("1. TWO POINTERS - Reverse Array")
print("-" * 50)

def reverse_array(arr):
    """Two pointers from start and end"""
    left, right = 0, len(arr) - 1
    
    while left < right:
        # Swap
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    
    return arr

arr = [1, 2, 3, 4, 5]
print(f"Original: {arr}")
print(f"Reversed: {reverse_array(arr)}")
print(f"Time: O(n), Space: O(1)\n")


# Pattern 2: Sliding Window
print("2. SLIDING WINDOW - Max Sum of Subarray Size k")
print("-" * 50)

def max_sum_subarray(arr, k):
    """Find maximum sum of subarray of size k"""
    if len(arr) < k:
        return None
    
    # Calculate sum of first window
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    # Slide the window
    for i in range(len(arr) - k):
        # Remove leftmost element, add new rightmost element
        window_sum = window_sum - arr[i] + arr[i + k]
        max_sum = max(max_sum, window_sum)
    
    return max_sum

arr = [1, 4, 2, 10, 2, 3, 1, 0, 20]
k = 4
result = max_sum_subarray(arr, k)
print(f"Array: {arr}")
print(f"Max sum of subarray size {k}: {result}")
print(f"Time: O(n), Space: O(1)\n")


# Pattern 3: Prefix Sum
print("3. PREFIX SUM - Range Sum Query")
print("-" * 50)

def build_prefix_sum(arr):
    """Build prefix sum array"""
    prefix = [0] * (len(arr) + 1)
    
    for i in range(len(arr)):
        prefix[i + 1] = prefix[i] + arr[i]
    
    return prefix

def range_sum(prefix, left, right):
    """Query sum of elements from index left to right"""
    return prefix[right + 1] - prefix[left]

arr = [1, 2, 3, 4, 5, 6]
prefix = build_prefix_sum(arr)
print(f"Array: {arr}")
print(f"Prefix sum: {prefix}")
print(f"Sum from index 1 to 4: {range_sum(prefix, 1, 4)}")
print(f"Preprocessing: O(n), Query: O(1)\n")


# Pattern 4: HashMap - Two Sum
print("4. HASHMAP - Two Sum")
print("-" * 50)

def two_sum(arr, target):
    """Find two numbers that add up to target"""
    seen = {}
    
    for num in arr:
        complement = target - num
        
        if complement in seen:
            return [seen[complement], num]
        
        seen[num] = num
    
    return None

arr = [2, 7, 11, 15]
target = 9
result = two_sum(arr, target)
print(f"Array: {arr}, Target: {target}")
print(f"Two sum: {result}")
print(f"Time: O(n), Space: O(n)\n")


# Pattern 5: Sorting / Partition
print("5. PARTITION - Move Zeros to End")
print("-" * 50)

def move_zeros(arr):
    """Move all zeros to end while maintaining order"""
    write_index = 0
    
    # Move all non-zero elements to front
    for read_index in range(len(arr)):
        if arr[read_index] != 0:
            arr[write_index], arr[read_index] = arr[read_index], arr[write_index]
            write_index += 1
    
    return arr

arr = [0, 1, 0, 3, 12]
print(f"Original: {arr}")
print(f"After moving zeros: {move_zeros(arr)}")
print(f"Time: O(n), Space: O(1)\n")
````

````python
print("=== COMPLETE EXAMPLE: Interview Problem ===\n")
print("PROBLEM: Find All Triplets That Sum to Zero (3Sum)\n")
print("=" * 60)

print("\nApproach 1: Brute Force (Interview starter, SLOW)")
print("-" * 60)

def three_sum_brute(arr):
    """Brute force - check all combinations"""
    triplets = set()
    n = len(arr)
    
    # Three nested loops - O(n³)
    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                if arr[i] + arr[j] + arr[k] == 0:
                    # Sort to avoid duplicates like (1,2,-3) vs (-3,1,2)
                    triplets.add(tuple(sorted([arr[i], arr[j], arr[k]])))
    
    return list(triplets)

arr = [-1, 0, 1, 2, -1, -4]
result = three_sum_brute(arr)
print(f"Input: {arr}")
print(f"Triplets summing to 0: {result}")
print(f"Time: O(n³) - SLOW, Space: O(1)")
print(f"Verdict: Good starting point but TOO SLOW for interviews\n")


print("\nApproach 2: Optimized with Sorting + Two Pointers (RECOMMENDED)")
print("-" * 60)

def three_sum_optimized(arr):
    """Optimized - sort then use two pointers"""
    arr.sort()  # O(n log n)
    triplets = []
    n = len(arr)
    
    # Fix one element, use two pointers for other two
    for i in range(n - 2):
        # Skip duplicates
        if i > 0 and arr[i] == arr[i - 1]:
            continue
        
        # If all remaining numbers positive, can't sum to 0
        if arr[i] > 0:
            break
        
        # Two pointers approach
        left, right = i + 1, n - 1
        
        while left < right:
            total = arr[i] + arr[left] + arr[right]
            
            if total == 0:
                triplets.append([arr[i], arr[left], arr[right]])
                
                # Skip duplicates for left pointer
                while left < right and arr[left] == arr[left + 1]:
                    left += 1
                # Skip duplicates for right pointer
                while left < right and arr[right] == arr[right - 1]:
                    right -= 1
                
                left += 1
                right -= 1
            
            elif total < 0:
                left += 1  # Need larger sum
            else:
                right -= 1  # Need smaller sum
    
    return triplets

arr = [-1, 0, 1, 2, -1, -4]
result = three_sum_optimized(arr)
print(f"Input: {arr}")
print(f"Triplets summing to 0: {result}")
print(f"Time: O(n²) - Much better!")
print(f"Space: O(1) - No extra space needed")
print(f"Verdict: Interview GOLD STANDARD ✓\n")


print("\nWhy Approach 2 is Better:")
print("-" * 60)
print("""
1. Time: O(n²) vs O(n³) - 100x faster for n=100!
2. Handles duplicates correctly
3. Shows you know optimized techniques
4. Scales to n = 100,000+
5. Interviewer impressed ✓
""")


print("\nInterview Flow for this Problem:")
print("-" * 60)
print("""
Interviewer: "Find all triplets that sum to zero"

YOU: 
"Let me think out loud:

1. Brute force: Check all combinations
   - Time: O(n³), Space: O(1)
   - Simple but SLOW

2. Better approach: Sort + Two Pointers
   - Sort: O(n log n)
   - For each element, use two pointers: O(n²)
   - Total: O(n²), Space: O(1)
   - Skip duplicates to avoid repeated triplets

3. Edge cases to handle:
   - Duplicate elements
   - Early termination if element > 0
   - Duplicate pointers skipping

Let me code the optimized solution:"

[Shows code above]

"Time: O(n²), Space: O(1) excluding output.
This is optimal for this problem."

Interviewer: IMPRESSED! ✓
""")


print("\nCommon Mistakes on This Problem:")
print("-" * 60)
print("""
WRONG: Forgetting to handle duplicates
  → Results in duplicate triplets like [0,0,0], [0,0,0]

WRONG: Not sorting array first
  → Can't use two pointers effectively

WRONG: Modifying input unnecessarily
  → Always copy if you need to modify

WRONG: Starting left/right from wrong positions
  → Can access same element multiple times

RIGHT: Sort, fix one element, two pointers on rest, skip duplicates
""")

print("\n" + "=" * 60)
print("KEY TAKEAWAY: Master sorting + two pointers for O(n²) solution!")
print("=" * 60)
````

### The Golden Questions to Ask

1. **Can I modify the input array?**
   - If YES: In-place solutions (O(1) space)
   - If NO: Extra space okay (O(n) space)

2. **Is the array sorted?**
   - If YES: Binary search or two pointers
   - If NO: Could sort it first (O(n log n))

3. **What about duplicates?**
   - Need to skip? Plan for it
   - Allowed? Handle in logic

4. **What are the constraints?**
   - Array size: If small, brute force okay
   - Value range: If limited, counting sort possible
   - Time limit: If tight, need O(n) or O(n log n)

### Time Complexity Hierarchy

```
O(1)           - Constant: Array access arr[i]
O(log n)       - Logarithmic: Binary search
O(n)           - Linear: Single loop
O(n log n)     - Linearithmic: Sorting, efficient searches
O(n²)          - Quadratic: Two nested loops, bubble sort
O(n³)          - Cubic: Three nested loops (avoid!)
O(2^n)         - Exponential: Backtracking all subsets
O(n!)          - Factorial: All permutations
```

**Interview Rule**: 
- n = 1,000: O(n²) acceptable
- n = 10,000: Need O(n log n) or better
- n = 100,000: Need O(n) or O(n log n)

### Space Complexity Hierarchy

```
O(1)       - Constant: In-place, no extra space
O(log n)   - Logarithmic: Recursion depth (binary search)
O(n)       - Linear: Extra array, HashMap
O(n²)      - Quadratic: 2D array, usually avoid
```

**Interview Rule**: 
- If no constraint: O(n) space is fine
- If asked "can you do better": Try for O(1) or O(log n)
- If "space limited": Must think in-place

```python
        if result == target:
```

### Code Template - Sliding Window

```python
def sliding_window_solution(arr, target):
    left = 0
    window_sum = 0
    results = []
    
    for right in range(len(arr)):
        # Expand window
        window_sum += arr[right]
        
        # Shrink window if needed
        while window_sum > target and left <= right:
            window_sum -= arr[left]
            left += 1
        
        # Record result if valid
        if window_sum == target:
            results.append((left, right))
    
    return results
```

### Code Template - Prefix Sum

```python
def prefix_sum_solution(arr):
    # Build prefix array
    prefix = [0] * (len(arr) + 1)
    for i in range(len(arr)):
        prefix[i + 1] = prefix[i] + arr[i]
    
    # Query range sum in O(1)
    def range_sum(left, right):
        return prefix[right + 1] - prefix[left]
    
    return range_sum
```

### Code Template - Binary Search

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:  # Note: <=
        mid = left + (right - left) // 2  # Avoid overflow
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1  # Search right half
        else:
            right = mid - 1  # Search left half
    
    return -1  # Not found
```

### Red Flags to Avoid

```
PROBLEM: "Brute force is O(n³)"
RED FLAG: Not looking for optimization
ACTION: Think two pointers, prefix sum, or sorting

PROBLEM: Array modification during iteration
RED FLAG: Index shifting issues
ACTION: Either iterate backward or use two pointers

PROBLEM: Off-by-one errors
RED FLAG: Loop bounds wrong
ACTION: Trace through small example

PROBLEM: Not handling empty array
RED FLAG: Will crash on edge case
ACTION: Add if not arr: return early

PROBLEM: Modifying original input
RED FLAG: Interviewer might care about this
```
