# Pattern: Sliding Window

## What is Sliding Window?

A **sliding window** is a technique where you maintain a contiguous subarray (window) of fixed or variable size and slide it through the array. As you move the window, you update calculations without recalculating from scratch.

## Core Idea

Instead of recalculating sum/count/property for every subarray:

```
Naive approach:     Check every possible subarray O(n²)
Sliding window:     Keep running calculation, add one element, 
                    remove one element O(n)
```

## Visual Example

```
Array: [1, 3, 2, 6, -1, 4, 1, 8]
Find: Maximum sum of subarray with size 3

Window size = 3

Step 1: [1, 3, 2]           → Sum = 6
        
Step 2:    [3, 2, 6]        → Sum = 11
           
Step 3:       [2, 6, -1]    → Sum = 7
              
Step 4:          [6, -1, 4] → Sum = 9
                 
Step 5:             [-1, 4, 1]  → Sum = 4
                    
Step 6:                [4, 1, 8]  → Sum = 13

ANSWER: Maximum sum = 13
```

## Two Types of Sliding Windows

### Type 1: Fixed Size Window

- Window size is constant (e.g., always 3)
- Easier to implement
- Examples: Max sum of k elements, average of k subarrays

### Type 2: Variable Size Window

- Window expands/contracts based on condition
- More complex but more powerful
- Examples: Longest substring without repeating, min window substring

---

## How Sliding Window Works

### The Key Concept: Two Pointers

A sliding window uses two pointers: `left` and `right`

```python
left = 0
right = 0

while right < n:
    # Expand window: add element at right
    window_sum += arr[right]
    
    # Shrink window if condition violated
    while condition_violated:
        window_sum -= arr[left]
        left += 1
    
    # Update answer with current window
    answer = max(answer, window_sum)
    
    right += 1
```

### Fixed Size vs Variable Size

```
FIXED SIZE:
┌─────────────────────────────────┐
│ Window size = k (always 3)      │
├─────────────────────────────────┤
│ Right moves: when window full   │
│ Left moves: automatically        │
│ Simpler to code                 │
└─────────────────────────────────┘

VARIABLE SIZE:
┌─────────────────────────────────┐
│ Window size = ? (expands/shrinks)│
├─────────────────────────────────┤
│ Right moves: always expanding   │
│ Left moves: when condition bad  │
│ More flexible, harder to code   │
└─────────────────────────────────┘
```

## When to Use Sliding Window

### Use When:

1. **Contiguous subarray/substring** problems
2. **Optimize from O(n²) to O(n)** - typical speedup!
3. **Fixed window size** - explicit k value
4. **Variable window** - "find longest/shortest" with condition
5. **No nested loops** - want O(n) instead of O(n²)

### Don't Use When:

1. **Array elements can be negative** (for fixed window min/max)
2. **Non-contiguous** elements matter
3. **Multiple arrays** to process together
4. **Need all subarrays** - sliding window finds just one

### Recognize Sliding Window Problems

Look for keywords in problem statement:

- "Longest" substring / subarray
- "Shortest" substring / subarray
- "Maximum" / "minimum" sum of subarray
- "Contains" or "without" (without repeating chars)
- "At most" / "At least" k elements
- "Substring" or "contiguous" subarray

## Time & Space Complexity

```
Sliding Window (General):
- Time: O(n) - each element visited at most twice
- Space: O(1) to O(k) - depends on what you store

Compare to Naive:
- Naive: O(n²) for checking all subarrays
- Sliding Window: O(n) - huge improvement!
```


#     Find maximum sum of contiguous subarray of size k

````python
print("Maximum Sum of Subarray with Fixed Size k")
print("=" * 70 + "\n")

def max_sum_subarray_k(arr, k):
    """
    Time: O(n)
    Space: O(1)
    """
    if len(arr) < k:
        return None
    
    # Calculate sum of first window
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    # Slide the window: remove left, add right
    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i - k]
        max_sum = max(max_sum, window_sum)
    
    return max_sum

# Test
arr = [1, 3, 2, 6, -1, 4, 1, 8]
k = 3

result = max_sum_subarray_k(arr, k)
print(f"Array: {arr}")
print(f"Window size: {k}")
print(f"Maximum sum: {result}")
print()

# Show all windows
print("All windows:")
for i in range(len(arr) - k + 1):
    window = arr[i:i+k]
    s = sum(window)
    print(f"  Window {i}: {window} → Sum = {s}")

print("\n")
````

#  Find length of longest substring without repeating characters

````python
def lengthOfLongestSubstring(s):
    seen = {}           # Line 1: Track last seen index of each character
    left = 0            # Line 2: Left boundary of window
    best = 0            # Line 3: Best answer found so far
    
    for right, ch in enumerate(s):  # Line 4: Iterate through string
        # Line 5-6: If character seen before AND in current window
        if ch in seen and seen[ch] >= left:
            left = seen[ch] + 1      # Line 7: Move left pointer past duplicate
        
        seen[ch] = right             # Line 8: Update last seen index
        best = max(best, right - left + 1)  # Line 9: Update best length
    
    return best                      # Line 10: Return answer
print(lengthOfLongestSubstring("abcabcbb"))  
````
