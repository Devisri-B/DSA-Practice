# Pattern: Sliding Window

## What is Sliding Window?

A **sliding window** is a technique where you maintain a contiguous subarray (window) of fixed or variable size and slide it through the array. As you move the window, you update calculations without recalculating from scratch.

## Core Idea

Instead of recalculating sum/count/property for every subarray:

```
Naive approach:     Check every possible subarray O(n²)
Sliding window:     Keep running calculation, add one element, 
                    remove one element O(n)
```

## Visual Example

```
Array: [1, 3, 2, 6, -1, 4, 1, 8]
Find: Maximum sum of subarray with size 3

Window size = 3

Step 1: [1, 3, 2]           → Sum = 6
        
Step 2:    [3, 2, 6]        → Sum = 11
           
Step 3:       [2, 6, -1]    → Sum = 7
              
Step 4:          [6, -1, 4] → Sum = 9
                 
Step 5:             [-1, 4, 1]  → Sum = 4
                    
Step 6:                [4, 1, 8]  → Sum = 13

ANSWER: Maximum sum = 13
```

## Two Types of Sliding Windows

### Type 1: Fixed Size Window

- Window size is constant (e.g., always 3)
- Easier to implement
- Examples: Max sum of k elements, average of k subarrays

### Type 2: Variable Size Window

- Window expands/contracts based on condition
- More complex but more powerful
- Examples: Longest substring without repeating, min window substring

---

## How Sliding Window Works

### The Key Concept: Two Pointers

A sliding window uses two pointers: `left` and `right`

```python
left = 0
right = 0

while right < n:
    # Expand window: add element at right
    window_sum += arr[right]
    
    # Shrink window if condition violated
    while condition_violated:
        window_sum -= arr[left]
        left += 1
    
    # Update answer with current window
    answer = max(answer, window_sum)
    
    right += 1
```

### Fixed Size vs Variable Size

```
FIXED SIZE:
┌─────────────────────────────────┐
│ Window size = k (always 3)      │
├─────────────────────────────────┤
│ Right moves: when window full   │
│ Left moves: automatically        │
│ Simpler to code                 │
└─────────────────────────────────┘

VARIABLE SIZE:
┌─────────────────────────────────┐
│ Window size = ? (expands/shrinks)│
├─────────────────────────────────┤
│ Right moves: always expanding   │
│ Left moves: when condition bad  │
│ More flexible, harder to code   │
└─────────────────────────────────┘
```

## When to Use Sliding Window

### Use When:

1. **Contiguous subarray/substring** problems
2. **Optimize from O(n²) to O(n)** - typical speedup!
3. **Fixed window size** - explicit k value
4. **Variable window** - "find longest/shortest" with condition
5. **No nested loops** - want O(n) instead of O(n²)

### Don't Use When:

1. **Array elements can be negative** (for fixed window min/max)
2. **Non-contiguous** elements matter
3. **Multiple arrays** to process together
4. **Need all subarrays** - sliding window finds just one

### Recognize Sliding Window Problems

Look for keywords in problem statement:

- "Longest" substring / subarray
- "Shortest" substring / subarray
- "Maximum" / "minimum" sum of subarray
- "Contains" or "without" (without repeating chars)
- "At most" / "At least" k elements
- "Substring" or "contiguous" subarray

## Time & Space Complexity

```
Sliding Window (General):
- Time: O(n) - each element visited at most twice
- Space: O(1) to O(k) - depends on what you store

Compare to Naive:
- Naive: O(n²) for checking all subarrays
- Sliding Window: O(n) - huge improvement!
```


#     Find maximum sum of contiguous subarray of size k

````python
print("Maximum Sum of Subarray with Fixed Size k")
print("=" * 70 + "\n")

def max_sum_subarray_k(arr, k):
    """
    Time: O(n)
    Space: O(1)
    """
    if len(arr) < k:
        return None
    
    # Calculate sum of first window
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    # Slide the window: remove left, add right
    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i - k]
        max_sum = max(max_sum, window_sum)
    
    return max_sum

# Test
arr = [1, 3, 2, 6, -1, 4, 1, 8]
k = 3

result = max_sum_subarray_k(arr, k)
print(f"Array: {arr}")
print(f"Window size: {k}")
print(f"Maximum sum: {result}")
print()

# Show all windows
print("All windows:")
for i in range(len(arr) - k + 1):
    window = arr[i:i+k]
    s = sum(window)
    print(f"  Window {i}: {window} → Sum = {s}")

print("\n")
````

#  Find length of longest substring without repeating characters

````python
def lengthOfLongestSubstring(s):
    seen = {}           # Line 1: Track last seen index of each character
    left = 0            # Line 2: Left boundary of window
    best = 0            # Line 3: Best answer found so far
    
    for right, ch in enumerate(s):  # Line 4: Iterate through string
        # Line 5-6: If character seen before AND in current window
        if ch in seen and seen[ch] >= left:
            left = seen[ch] + 1      # Line 7: Move left pointer past duplicate
        
        seen[ch] = right             # Line 8: Update last seen index
        best = max(best, right - left + 1)  # Line 9: Update best length
    
    return best                      # Line 10: Return answer
print(lengthOfLongestSubstring("abcabcbb"))  
````

## Time & Space Complexity

```
Time Complexity: O(n * m)
  n = len(source)
  m = len(pattern)
  
  - Outer loop: n - m + 1 iterations
  - Inner loop: m iterations
  - Total: (n - m + 1) * m = O(n*m)
  
  Worst case: pattern length = n/2
  Then: O(n * n/2) = O(n²)

Space Complexity: O(1)
  - vowels set: O(26) = O(1) constant
  - Other variables: O(1)
```

Actually YES! We can use **true sliding window** to make it O(n):

```python
def solution_optimized(pattern, source):

    plen = len(pattern)
    
    # Pre-compute match for first window
    def check_match(start):
        for i in range(plen):
            is_vowel = source[start + i] in vowels
            if (pattern[i] == "0" and is_vowel) or (pattern[i] == "1" and not is_vowel):
                return False
        return True

    for start in range(len(source) - plen + 1):
        if check_match(start):
```

The current implementation is actually already O(n*m) which is fine for most cases. A true O(n) solution isn't possible here because we need to recheck the entire window pattern each time (there's no way to incrementally update the match without checking all positions).

## The Boundary Formula Explained

### Why `range(len(source) - plen + 1)`?

```
Let's say:
- source = "abcde" (length 5)
- pattern = "010" (length 3)

We want substrings of length 3:
- [0:3] = "abc"    ✓ index 0
- [1:4] = "bcd"    ✓ index 1
- [2:5] = "cde"    ✓ index 2
- [3:6] = ???      ✗ Out of bounds! (only goes to index 4)

So valid start positions: 0, 1, 2
That's range(5 - 3 + 1) = range(3) ✓

Formula: range(len(source) - plen + 1)
        range(5 - 3 + 1) = range(3)
```

### Why `+ 1`?

```
Without +1: range(5 - 3) = range(2) = [0, 1]
  Missing position 2!

With +1: range(5 - 3 + 1) = range(3) = [0, 1, 2]
  Correct! We check all valid windows
```




## The Big Picture: Fixed Size Sliding Window Pattern

### When to Use This Pattern

```
Use Fixed Size Sliding Window when:

1. Window size is constant and known
2. Need to check/process all windows of that size
3. Want to count/find patterns in fixed-length windows
4. Problem mentions "substring of length k"

Don't use when:

1. Window size varies based on condition
2. Only need one window, not all windows
3. Problem asks about variable-length substrings
4. Efficiency doesn't matter (could use brute force)
```

### Comparison: Sliding Window vs Brute Force

```
BRUTE FORCE (Naive):
Time: O(n * m)
- n iterations (starting positions)
- m iterations (checking each position)
- Actually same complexity as sliding window!

So why use sliding window then?
- Cleaner code
- More efficient in practice (break early)
- Sets pattern for harder problems
- Easier to optimize later
```

## Key Insights

### 1. The Boundary Formula is Critical

```python
# WRONG: range(len(source) - plen)
# This stops one window too early!

# RIGHT: range(len(source) - plen + 1)
# This includes all valid starting positions
```

**Why?** You need the `+1` because:
- Last window starts at index: `len(source) - plen`
- But range is exclusive of the end
- So range needs to go to: `len(source) - plen + 1`

### 2. Early Termination Saves Time

```python
# Without break: Always checks all m characters
for i in range(plen):
    # check...
# O(m) per window

# With break: Can exit early on mismatch
for i in range(plen):
    # check...
    if mismatch:
        break  # Exit loop early
# O(1) to O(m) per window, avg much less
```

### 3. Data Structure Matters

```python
# Slow: Check membership in string
if ch in "aeiou":  # O(5) = O(1) constant but with coefficient

# Fast: Check membership in set
if ch in vowels:  # O(1) true O(1)

# When checking millions of chars, this matters!
```

### 4. Logic Verification is Essential

For pattern matching, always verify:
```
Pattern means:   '0' = consonant, '1' = vowel

Then logic should be:
- "If pattern[i] == '0' and char IS vowel" → MISMATCH
- "If pattern[i] == '1' and char is NOT vowel" → MISMATCH

NOT the inverted version from the original code!
```
