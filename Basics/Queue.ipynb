## Section 2: Queue Data Structure Basics

### What is a Queue?

A **queue** is a linear data structure that follows the FIFO (First In First Out) principle:
- The element that enters FIRST will exit FIRST

### Key Terminology

- **Front/Head**: Position where elements are removed (dequeued)
- **Rear/Back/Tail**: Position where elements are added (enqueued)
- **Enqueue**: Add element to the rear
- **Dequeue**: Remove and return front element
- **Peek**: View front element without removing it
- **isEmpty**: Check if queue contains no elements

### Visual Representation

```
Enqueue 1, 2, 3 (left to right):

Front       Rear
  ↓          ↓
[1] ← [2] ← [3]

Dequeue removes 1 (First In)
Dequeue removes 2
Dequeue removes 3 (First Out)
```

### Queue vs Stack (Key Difference)

| Data Structure | Order | Real-World Example |
|---|---|---|
| **Stack (LIFO)** | Last element added exits first | Undo button (last action undone first) |
| **Queue (FIFO)** | First element added exits first | Coffee shop line (first person served first) |

### Common Real-World Use Cases

1. **Customer Service Lines**: First person in line is first to be served
2. **Print Queue**: Documents printed in order they were submitted
3. **Task Scheduler**: Tasks executed in order they were added
4. **Breadth-First Search (BFS)**: Traversing trees/graphs level by level
5. **Network Packets**: Packets sent in order (TCP/IP)

## Section 3: Implement Queue Using Lists

### Approach: List-Based Implementation

Using Python lists with `append()` to add and `pop(0)` to remove from front:
- `append()`: Adds element at end (rear) - O(1)
- `pop(0)`: Removes and returns first element (front) - O(n) SLOW!

### The Problem with Lists

```
Removing from front using pop(0) requires COPYING all elements:

Before: [1, 2, 3, 4, 5]
Remove first element: O(n) operation
        ↓
After:  [2, 3, 4, 5]
        (All elements shift left - expensive!)
```

### Time and Space Complexity

- **Enqueue (append)**: O(1) amortized
- **Dequeue (pop(0))**: O(n) SLOW - must copy all elements
- **Peek**: O(1)
- **isEmpty**: O(1)
- **Space**: O(n) where n is number of elements

### Why Lists Are Bad for Queues

If you process 1000 items, you do 1000 O(n) operations = O(n²) total - VERY SLOW!

````python
class QueueList:
    """Queue implementation using Python list (NOT RECOMMENDED for large queues)"""
    
    def __init__(self):
        """Initialize empty queue"""
        self.items = []
    
    def enqueue(self, item):
        """Add element to rear (end) - O(1)"""
        self.items.append(item)
    
    def dequeue(self):
        """Remove and return front element - O(n) SLOW!"""
        if self.is_empty():
            raise IndexError("Dequeue from empty queue")
        return self.items.pop(0)  # pop(0) is O(n) - must copy all elements
    
    def peek(self):
        """View front element without removing - O(1)"""
        if self.is_empty():
            raise IndexError("Peek on empty queue")
        return self.items[0]
    
    def is_empty(self):
        """Check if queue is empty - O(1)"""
        return len(self.items) == 0
    
    def size(self):
        """Return number of elements - O(1)"""
        return len(self.items)
    
    def __str__(self):
        """String representation (front to rear)"""
        return f"Queue({self.items})" if self.items else "Queue([])"


# Test the List-based Queue
print("=== Testing List-Based Queue (SLOW for dequeue) ===\n")
queue = QueueList()

# Enqueue elements
print("Enqueuing 10, 20, 30:")
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print(f"Queue: {queue}")
print(f"Size: {queue.size()}")

# Peek
print(f"\nPeek (view front): {queue.peek()}")

# Dequeue
print(f"\nDequeuing: {queue.dequeue()}")  # Removes 10
print(f"Queue after dequeue: {queue}")

print(f"\nDequeuing: {queue.dequeue()}")  # Removes 20
print(f"Queue after dequeue: {queue}")

# Check empty
print(f"\nIs empty? {queue.is_empty()}")

# Dequeue remaining
print(f"\nDequeuing: {queue.dequeue()}")  # Removes 30
print(f"Is empty? {queue.is_empty()}")

# Try to dequeue from empty
try:
    queue.dequeue()
except IndexError as e:
    print(f"Error caught: {e}")
````

## Section 4: Implement Queue Using Collections.deque (RECOMMENDED)

### Why Use Deque?

Deque (double-ended queue) is **optimized for adding and removing from BOTH ends**:
- `append()`: Add to rear - O(1)
- `popleft()`: Remove from front - O(1) FAST!
- No element shifting needed

### The Architecture Advantage

```
Deque popleft (FAST):
Block1: [1, 2]    Block2: [3, 4, 5]
 ↓ Just move pointer
Block2: [3, 4, 5]  - O(1) operation
```

### Time and Space Complexity

| Operation | List | Deque | Difference |
|---|---|---|---|
| enqueue (append) | O(1) amortized | O(1) guaranteed | Same |
| dequeue (popleft) | O(n) | O(1) | **Deque 100x faster!** |
| peek | O(1) | O(1) | Same |
| isEmpty | O(1) | O(1) | Same |

### When to Use Deque vs List

- **Small queue (< 100 items)**: List is fine
- **Large queue or frequent dequeues**: Use deque
- **Real-time systems (gaming, trading)**: Always use deque (predictable O(1))

````python
class QueueDeque:
    """Queue implementation using collections.deque (RECOMMENDED)"""
    
    def __init__(self):
        """Initialize empty queue with deque"""
        self.items = deque()
    
    def enqueue(self, item):
        """Add element to rear - O(1)"""
        self.items.append(item)
    
    def dequeue(self):
        """Remove and return front element - O(1)"""
        if self.is_empty():
            raise IndexError("Dequeue from empty queue")
        return self.items.popleft()  # popleft() is O(1) - FAST!
    
    def peek(self):
        """View front element without removing - O(1)"""
        if self.is_empty():
            raise IndexError("Peek on empty queue")
        return self.items[0]
    
    def is_empty(self):
        """Check if queue is empty - O(1)"""
        return len(self.items) == 0
    
    def size(self):
        """Return number of elements - O(1)"""
        return len(self.items)
    
    def __str__(self):
        """String representation (front to rear)"""
        return f"Queue({list(self.items)})" if self.items else "Queue([])"


# Test Deque-Based Queue
print("\n=== Testing Deque-Based Queue (FAST - RECOMMENDED) ===\n")
queue_deque = QueueDeque()

# Enqueue elements
print("Enqueuing 'Alice', 'Bob', 'Charlie':")
queue_deque.enqueue('Alice')
queue_deque.enqueue('Bob')
queue_deque.enqueue('Charlie')
print(f"Queue: {queue_deque}")
print(f"Size: {queue_deque.size()}")

# Peek
print(f"\nPeek (view front): {queue_deque.peek()}")

# Dequeue all
print(f"\nDequeuing in order: ", end="")
while not queue_deque.is_empty():
    print(queue_deque.dequeue(), end=" ")
print("\n")

print(f"Queue after all dequeues: {queue_deque}")
print(f"Is empty? {queue_deque.is_empty()}")
````

## Section 5: Common Mistakes to Avoid

1. **Using List for Large Queues**: pop(0) is O(n) - use deque instead
2. **Forgetting to Check Empty**: Always check before dequeue/peek
3. **Confusing with Stack**: Stack is LIFO (last out), Queue is FIFO (first out)
4. **Incorrect Method Names**: `popleft()` for deque, NOT `pop(0)` for lists

## Section 6: Stack-Based Queue Implementation

### Concept: Using Two Stacks to Create a Queue

You can implement a queue using **two stacks**! 
### How It Works

```
Stack 1 (Input):     Stack 2 (Output):
Push 1, 2, 3         Empty initially

When dequeuing:
1. Pop from Stack 2 (returns 1 - FIFO!)
2. If Stack 2 empty, reverse Stack 1 into Stack 2
3. Then pop from Stack 2

Magic: Even though stacks are LIFO, two stacks together create FIFO!
```

### Visual Example

```
Enqueue 1, 2, 3 (push to Stack 1):
Stack 1:  [3, 2, 1] ← top
Stack 2:  []

Dequeue:
- Stack 2 is empty, so reverse Stack 1
- Pop from Stack 1: 1, 2, 3 → push to Stack 2
Stack 1:  []
Stack 2:  [1, 2, 3] ← top

Dequeue (returns 3, then 2, then 1):
- Pop from Stack 2: returns 1 (FIFO!) ✓
- Pop from Stack 2: returns 2 ✓
- Pop from Stack 2: returns 3 ✓
```

### Amortized Complexity

| Operation | Time | Why? |
|-----------|------|------|
| enqueue | O(1) | Just push to Stack 1 |
| dequeue | O(1) amortized | Most pops are O(1), reversal happens once per element |
| peek | O(1) amortized | Same as dequeue |

### When Each Operation is Called

- **Enqueue**: O(1) always - just push to input stack
- **First dequeue after many enqueues**: O(n) - need to reverse input stack
- **Subsequent dequeues**: O(1) - just pop from output stack
- **Average**: O(1) amortized

````python
class QueueWithTwoStacks:
    """Queue implemented using two stacks"""
    
    def __init__(self):
        self.input_stack = []    # For enqueue operations
        self.output_stack = []   # For dequeue operations
    
    def enqueue(self, item):
        """Add element - O(1)"""
        self.input_stack.append(item)
        print(f"Enqueued {item}: input_stack = {self.input_stack}")
    
    def _reverse_if_needed(self):
        """Reverse input stack to output stack if output is empty"""
        if len(self.output_stack) == 0:
            print("Output stack empty, reversing input stack...")
            while self.input_stack:
                item = self.input_stack.pop()
                self.output_stack.append(item)
    
    def dequeue(self):
        """Remove and return front element - O(1) amortized"""
        self._reverse_if_needed()
        
        if len(self.output_stack) == 0:
            raise IndexError("Dequeue from empty queue")
        
        item = self.output_stack.pop()
        print(f"Dequeued {item}: output_stack = {self.output_stack}")
        return item
    
    def peek(self):
        """View front element - O(1) amortized"""
        self._reverse_if_needed()
        
        if len(self.output_stack) == 0:
            raise IndexError("Peek on empty queue")
        
        return self.output_stack[-1]
    
    def is_empty(self):
        """Check if queue is empty - O(1)"""
        return len(self.input_stack) == 0 and len(self.output_stack) == 0
    
    def __str__(self):
        """String representation"""
        return f"Queue(input={self.input_stack}, output={self.output_stack})"


# Test Stack-Based Queue
print("=== Testing Stack-Based Queue ===\n")
queue_stacks = QueueWithTwoStacks()

print("Enqueuing 10, 20, 30:")
queue_stacks.enqueue(10)
queue_stacks.enqueue(20)
queue_stacks.enqueue(30)

print(f"\nQueue state: {queue_stacks}\n")

print("Dequeuing:")
queue_stacks.dequeue()
queue_stacks.dequeue()

print(f"\nQueue state: {queue_stacks}\n")

print("Dequeuing last element:")
queue_stacks.dequeue()

print(f"\nQueue state: {queue_stacks}")
print(f"Is empty? {queue_stacks.is_empty()}")
````

## Section 7: Node-Based Queue (Linked List Implementation)

### Concept: Using Nodes and Pointers

A **node-based queue** uses individual nodes connected by pointers, similar to a linked list. This is the most elegant implementation - pure FIFO with O(1) operations.

### How It Works

```
Each node contains:
- data: the value
- next: pointer to next node

Queue with nodes:
Head → [1|next] → [2|next] → [3|next] → None
↑                                         ↑
Front (dequeue here)                Rear (enqueue here)

Enqueue 4:
Head → [1|next] → [2|next] → [3|next] → [4|None]
       ↑                                   ↑
       Dequeue                            Enqueue

Dequeue (returns 1):
      Head → [2|next] → [3|next] → [4|None]
             ↑                       ↑
             New front              Rear
```

### Architecture Advantage

```
Deque (dynamic array):
Block1: [1, 2]    Block2: [3, 4, 5]
- Still has overhead
- Amortized complexity

Linked List (nodes):
[1|→] → [2|→] → [3|→] → [4|→] → None
- Pure O(1) guaranteed
- No copying, no overhead
```

### Time Complexity

| Operation | Complexity | Why? |
|-----------|-----------|------|
| enqueue | O(1) | Just create node and link it |
| dequeue | O(1) | Just remove head pointer |
| peek | O(1) | Just read head value |
| isEmpty | O(1) | Check if head is None |
| Space | O(n) | One node per element |

### When to Use Node-Based Queue

- **Graph algorithms (BFS)**: Predictable performance
- **Real-time systems**: No garbage collection pauses
- **Interview questions**: Shows deep understanding
- **Learning**: Best way to understand how queues work internally

````python
class Node:
    """Node for linked list-based queue"""
    def __init__(self, data):
        self.data = data
        self.next = None


class QueueWithNodes:
    """Queue implemented using nodes and pointers (linked list)"""
    
    def __init__(self):
        self.head = None  # Front of queue (dequeue here)
        self.tail = None  # Rear of queue (enqueue here)
        self._size = 0
    
    def enqueue(self, item):
        """Add element to rear - O(1)"""
        new_node = Node(item)
        
        if self.is_empty():
            # Queue was empty, new node is both head and tail
            self.head = new_node
            self.tail = new_node
        else:
            # Link new node to end
            self.tail.next = new_node
            self.tail = new_node
        
        self._size += 1
        print(f"Enqueued {item}")
    
    def dequeue(self):
        """Remove and return front element - O(1)"""
        if self.is_empty():
            raise IndexError("Dequeue from empty queue")
        
        # Get value from head
        item = self.head.data
        
        # Move head to next node
        self.head = self.head.next
        self._size -= 1
        
        # If queue becomes empty, update tail too
        if self._size == 0:
            self.tail = None
        
        print(f"Dequeued {item}")
        return item
    
    def peek(self):
        """View front element - O(1)"""
        if self.is_empty():
            raise IndexError("Peek on empty queue")
        
        return self.head.data
    
    def is_empty(self):
        """Check if queue is empty - O(1)"""
        return self.head is None
    
    def size(self):
        """Return number of elements - O(1)"""
        return self._size
    
    def _print_queue(self):
        """Helper to visualize queue"""
        elements = []
        current = self.head
        while current:
            elements.append(str(current.data))
            current = current.next
        
        print(f"Queue: {' → '.join(elements) if elements else 'Empty'} → None")


# Test Node-Based Queue
print("=== Testing Node-Based Queue (Linked List) ===\n")
queue_nodes = QueueWithNodes()

print("Enqueuing 10, 20, 30:")
queue_nodes.enqueue(10)
queue_nodes.enqueue(20)
queue_nodes.enqueue(30)

queue_nodes._print_queue()
print(f"Size: {queue_nodes.size()}\n")

print("Peek (view front):", queue_nodes.peek())
print()

print("Dequeuing:")
queue_nodes.dequeue()
queue_nodes._print_queue()
queue_nodes.dequeue()
queue_nodes._print_queue()

print(f"Size: {queue_nodes.size()}\n")

print("Dequeuing last element:")
queue_nodes.dequeue()
queue_nodes._print_queue()
print(f"Is empty? {queue_nodes.is_empty()}")
````
