## Section 2: Stack Data Structure Basics

### What is a Stack?

A **stack** is a linear data structure that follows the LIFO (Last In First Out) principle:
- The element that enters LAST will exit FIRST
- Think of a stack of plates: you add plates on top and remove from the top

### Key Terminology

- **Top**: Position where elements are added and removed
- **Base/Bottom**: The bottom of the stack
- **Push**: Add element to the top
- **Pop**: Remove and return top element
- **Peek**: View top element without removing it
- **isEmpty**: Check if stack contains no elements

### Visual Representation

```
Push 1, 2, 3:

        3  <- Top (next to pop)
        2
        1
     (Base)

Pop returns 3 (Last In)
Pop returns 2
Pop returns 1 (First Out)
```

### Stack vs Queue (Key Difference)

- **Stack (LIFO)**: Last element added is first to be removed
- **Queue (FIFO)**: First element added is first to be removed (opposite order)

### Common Real-World Use Cases

1. **Browser Back Button**: Each page you visit is pushed; clicking back pops the page
2. **Undo/Redo**: Each action is pushed; undo pops it
3. **Function Call Stack**: When functions call each other, they use a stack (internally)
4. **Expression Evaluation**: Checking balanced parentheses, converting infix to postfix
5. **Depth-First Search (DFS)**: Traversing trees/graphs
6. **Backtracking**: Finding solutions in puzzles, maze solving

## Section 3: Implement Stack Using Lists

### Approach: List-Based Implementation

Using Python lists with `append()` and `pop()` methods:
- `append()`: Adds element at the end (O(1) amortized)
- `pop()`: Removes and returns last element (O(1))

### Time and Space Complexity

- **Push (append)**: O(1) amortized
- **Pop**: O(1)
- **Peek**: O(1)
- **isEmpty**: O(1)
- **Space**: O(n) where n is number of elements

### Important Note: What is "Amortized O(1)"?

"Amortized" means most operations are O(1), but occasionally ONE operation might be O(n):
- List stores elements in a **contiguous memory block** (dynamic array)
- When you keep appending and the block gets full, Python must:
  1. Allocate a BIGGER block of memory (usually 1.125x larger)
  2. COPY all existing elements to the new block
  3. Now you have space for more appends
- This happens rarely, so we say it's "amortized O(1)" - on average O(1), but with occasional pauses

Visual:
```
List internal storage:

append(): [1, 2, 3, _, _]        O(1) fast, just add
append(): [1, 2, 3, 4, _]        O(1) fast, just add
append(): FULL! [1, 2, 3, 4, 5]  O(1) fast, just add
append(): FULL! Need to grow!    O(n) slow! Copy all elements
          Allocate [_, _, _, _, _, _, _, _]
          Copy: [1, 2, 3, 4, 5, _, _, _]
```

````python
class StackList:
    """Stack implementation using Python list"""
    
    def __init__(self):
        """Initialize empty stack"""
        self.items = []
    
    def push(self, item):
        """Add element to top of stack - O(1)"""
        self.items.append(item)
    
    def pop(self):
        """Remove and return top element - O(1)"""
        if self.is_empty():
            raise IndexError("Pop from empty stack")
        return self.items.pop()
    
    def peek(self):
        """View top element without removing - O(1)"""
        if self.is_empty():
            raise IndexError("Peek on empty stack")
        return self.items[-1]
    
    def is_empty(self):
        """Check if stack is empty - O(1)"""
        return len(self.items) == 0
    
    def size(self):
        """Return number of elements - O(1)"""
        return len(self.items)
    
    def __str__(self):
        """String representation of stack (top to bottom)"""
        return f"Stack({self.items[::-1]})" if self.items else "Stack([])"


# Test the Stack implementation
print("=== Testing List-Based Stack ===")
stack = StackList()

# Push elements
print("\nPushing 10, 20, 30:")
stack.push(10)
stack.push(20)
stack.push(30)
print(f"Stack: {stack}")
print(f"Size: {stack.size()}")

# Peek
print(f"\nPeek (view top): {stack.peek()}")

# Pop
print(f"\nPopping: {stack.pop()}")
print(f"Stack after pop: {stack}")

print(f"\nPopping: {stack.pop()}")
print(f"Stack after pop: {stack}")

# Check empty
print(f"\nIs empty? {stack.is_empty()}")

# Pop remaining
print(f"\nPopping: {stack.pop()}")
print(f"Is empty? {stack.is_empty()}")

# Try to pop from empty (will raise error)
try:
    stack.pop()
except IndexError as e:
    print(f"Error caught: {e}")
````

## Section 4: Implement Stack Using Collections.deque

### Why Use Deque? The Architecture Difference

You're RIGHT that the operations look identical! But the **internal storage** is completely different:

### List (Dynamic Array) vs Deque (Linked Blocks)

**List Storage (what you see in memory):**
```
Single contiguous block:

Memory: [1][2][3][4][5][_][_][_]
        ^  ^  ^  ^  ^
        All stored together in one block
        
When full, must copy entire block to bigger space (expensive!)
```

**Deque Storage (what you see in memory):**
```
Multiple linked blocks (chunks):

Block 1:    Block 2:    Block 3:
[1][2][3]  [4][5][6]  [7][8][_]
   ↓          ↓           ↓
Connected via pointers (linked list of blocks)

When full, just create a new block and link it (cheap!)
No copying needed - new block is independent
```

### Why This Matters for Stacks

**List append() cost:**
- Most times: O(1) - just add to current space
- Sometimes: O(n) - when it needs to grow and copy everything

**Deque append() cost:**
- Always: O(1) guaranteed - just add to current block or create new one
- NO copying, NO "amortization" - truly constant time every single time

### The Real Difference (Operations Look Same, Internals Different)

```python
# These look identical:
list_stack = []
list_stack.append(1)      # Same operation
top = list_stack[-1]      # Same operation
list_stack.pop()          # Same operation

deque_stack = deque()
deque_stack.append(1)     # Same operation
top = deque_stack[-1]     # Same operation (also O(1))
deque_stack.pop()         # Same operation
```

### When Does This Become Critical?

**Small stack (< 1000 elements):** Probably doesn't matter much, both fast

**Large stack (> 10,000 elements):** 
- List has had to reallocate multiple times - occasional O(n) pauses
- Deque glides smoothly - every operation guaranteed O(1)

**Real-time system (trading app, game, medical device):**
- Deque predictable - no surprise pauses
- List unpredictable - might hiccup when reallocating

````python
class StackDeque:
    """Stack implementation using collections.deque (RECOMMENDED)"""
    
    def __init__(self):
        """Initialize empty stack with deque"""
        self.items = deque()
    
    def push(self, item):
        """Add element to top - O(1)"""
        self.items.append(item)
    
    def pop(self):
        """Remove and return top element - O(1)"""
        if self.is_empty():
            raise IndexError("Pop from empty stack")
        return self.items.pop()
    
    def peek(self):
        """View top element without removing - O(1)"""
        if self.is_empty():
            raise IndexError("Peek on empty stack")
        return self.items[-1]
    
    def is_empty(self):
        """Check if stack is empty - O(1)"""
        return len(self.items) == 0
    
    def size(self):
        """Return number of elements - O(1)"""
        return len(self.items)
    
    def __str__(self):
        """String representation"""
        return f"Stack({list(self.items)[::-1]})" if self.items else "Stack([])"


# Test Deque-Based Stack
print("\n=== Testing Deque-Based Stack (PREFERRED) ===")
stack_deque = StackDeque()

# Push elements
print("\nPushing 'A', 'B', 'C':")
stack_deque.push('A')
stack_deque.push('B')
stack_deque.push('C')
print(f"Stack: {stack_deque}")
print(f"Size: {stack_deque.size()}")

# Peek
print(f"\nPeek: {stack_deque.peek()}")

# Pop all
print(f"\nPopping in order: ", end="")
while not stack_deque.is_empty():
    print(stack_deque.pop(), end=" ")
print("\n")
````

## Section 6: Check if Stack is Empty - Error Handling

### Why Check isEmpty()?

Always check before popping or peeking to avoid crashes:
- Prevents IndexError exceptions
- Handles edge cases gracefully
- Makes code more robust and production-ready

### Best Practice: Guard Against Empty Stack

````python
# Method 1: Always check before popping
print("=== Safe Stack Operations ===\n")

stack = StackDeque()
stack.push(100)
stack.push(200)

print("Method 1: Check before popping")
if not stack.is_empty():
    value = stack.pop()
    print(f"Safely popped: {value}")
else:
    print("Stack is empty, skip pop")


# Method 2: Using try-except (alternative)
print("\n\nMethod 2: Using try-except (alternative)")
stack = StackDeque()

try:
    top_value = stack.peek()
    print(f"Top value: {top_value}")
except IndexError as e:
    print(f"Error handled: {e}")


# Method 3: Get with default value (practical pattern)
print("\n\nMethod 3: Safe pop with default value")

def safe_pop(stack, default=None):
    """Pop safely, return default if empty"""
    return stack.pop() if not stack.is_empty() else default

result = safe_pop(stack)
print(f"Result: {result} (default value since stack empty)")



# Method 4: Size check
print("\n\nMethod 4: Check size")
stack = StackDeque()
print(f"Stack size: {stack.size()}")
print(f"Is empty: {stack.size() == 0}")

````
