# C++

## Including Libraries

Libraries provide pre-built functions and tools.

- `#include<iostream>` - used for input/output
- `#include<math.h>` - used for mathematical functions

Every C++ program needs a main() function:

```cpp
#include<iostream>
int main() {
    // Your code here
    return 0;
}
```

## Output with cout

To print output, use `cout` from the iostream library. Since it belongs to the standard namespace, we write `std::cout`.

```cpp
#include<iostream>
int main() {
    std::cout << "Hey, Devi!";
    return 0;
}
```

Output: `Hey, Devi!`

## Printing on Multiple Lines

If you write cout statements repeatedly, they print on the same line:

```cpp
#include<iostream>
int main() {
    std::cout << "Hey, Devi!";
    std::cout << "Hey, Devi!";
    return 0;
}
```

Output: `Hey, Devi! Hey, Devi!`

### Using Newline Characters

Use `\n` (newline character) or `std::endl` to print on new lines:

```cpp
#include<iostream>
int main() {
    std::cout << "Hey, Devi!" << "\n";
    std::cout << "Hey, Devi!";
    return 0;
}
```

Output:
```
Hey, Devi!
Hey, Devi!
```

### Difference between \n and std::endl

- `\n` - inserts a new line (faster, commonly used)
- `std::endl` - inserts a new line and flushes the output buffer (slower)

```cpp
#include<iostream>
int main() {
    std::cout << "Hey, Devi!" << std::endl;
    std::cout << "Hey, Devi!";
    return 0;
}
```

## Using namespace std

Writing `using namespace std;` removes the need to prefix `std::`. This makes code cleaner but can cause naming conflicts in large projects.

```cpp
#include<iostream>
using namespace std;
int main() {
    cout << "Hey, Devi!" << endl;
    cout << "Hey, Devi!";
    return 0;
}
```

## Comments

Comments are used to explain code and make it more readable. They are ignored by the compiler.

- Single-line comment: `//`
- Multi-line comment: `/* ... */`

```cpp
// This is a single-line comment
int x = 5; // Comments can be inline

/* This is a multi-line comment
   that spans multiple lines */
int y = 10;
```

## Variables and Data Types

Variables store data values. Common data types in C++:

- `int` - integer (whole numbers)
- `float` - floating-point number (decimal numbers)
- `double` - floating-point number (more precision than float)
- `char` - single character
- `string` - text/sequence of characters
- `bool` - boolean (true or false)

```cpp
#include<iostream>
using namespace std;
int main() {
    int age = 25;           // integer
    float height = 5.9;     // floating-point
    char grade = 'A';       // character
    string name = "Devi";   // string
    bool isStudent = true;  // boolean
    
    cout << "Name: " << name << endl;
    cout << "Age: " << age << endl;
    
    return 0;
}
```

Output:
```
Name: Devi
Age: 25
```

## Taking User Input with cin

`cin` is used to take input from the user.

```cpp
#include<iostream>
using namespace std;
int main() {
    int x;
    cin >> x;
    cout << "Value of x: " << x;
    return 0;
}
```

Input: `10`

Output: `Value of x: 10`

### Multiple Inputs

```cpp
#include<iostream>
using namespace std;
int main() {
    int x, y;
    cin >> x >> y;
    cout << "Value of x: " << x << " and y: " << y;
    return 0;
}
```

Input: `10 20`

Output: `Value of x: 10 and y: 20`

### Taking String Input with getline

For reading strings with spaces, use `getline()`:

```cpp
#include<iostream>
using namespace std;
int main() {
    string fullName;
    getline(cin, fullName);  // Reads entire line including spaces
    cout << "Full Name: " << fullName;
    return 0;
}
```

Input: `David Christian`

Output: `Full Name: David Cristian`

## Shortcut: bits/stdc++.h

Instead of including libraries one by one, you can use:

```cpp
#include<bits/stdc++.h>
```

- Includes almost all standard libraries at once
- Useful in competitive programming
- NOT recommended for production code (due to compile-time overhead)

```cpp
#include<bits/stdc++.h>
using namespace std;
int main() {
    int x, y;
    cin >> x >> y;
    cout << "Value of x: " << x << " and y: " << y;
    return 0;
}
```

## Key Takeaways

- Every C++ program requires a `main()` function which returns an int
- Use `#include` to import libraries
- Use `cout` for output and `cin` for input
- Use `using namespace std;` to simplify code (but avoid in large projects)
- Comments make code more readable
- Use `\n` for better performance than `std::endl`
- For string input with spaces, use `getline()`

# Arrays

## What is an Array?

An array is a collection of elements of the same type stored in contiguous memory locations. Each element can be accessed using an index starting from 0.

## Creating an Array

### C++

**Syntax:**
```cpp
Data_type array_name[array_size];
```

**Example:**
```cpp
int myArray[8];  // Array of 8 integers
```

- `int` - Data type
- `myArray` - Name of array
- `[8]` - Size of array (fixed)

When you define the size of the array, all memory space is reserved instantly in memory.

**Initialization:**
```cpp
int myArray[5] = {10, 20, 30, 40, 50};  // Initialize with values
int arr[] = {1, 2, 3, 4};                // Size is automatically determined
```

### Python

Python doesn't have traditional fixed-size arrays. Use **lists** instead, which are dynamic and flexible:

**Syntax:**
```python
array_name = [element1, element2, element3]
```

**Examples:**
```python
myList = [10, 20, 30, 40, 50]           # Initialize with values
myList = []                              # Empty list
myList = [0] * 5                         # [0, 0, 0, 0, 0] - list of 5 zeros
myList = list(range(5))                  # [0, 1, 2, 3, 4]
```

**Using NumPy for Fixed Arrays (when needed):**
```python
import numpy as np
myArray = np.zeros(5)      # Array of 5 zeros
myArray = np.array([1, 2, 3, 4, 5])
```

**Key Difference:**
- C++ arrays are fixed-size and allocated at compile time
- Python lists are dynamic and can grow/shrink at runtime

## Accessing Array Elements

### C++

```cpp
int myArray[] = {10, 20, 30, 40, 50};
cout << myArray[0];   // Output: 10 (first element)
cout << myArray[4];   // Output: 50 (last element)
```

### Python

```python
myList = [10, 20, 30, 40, 50]
print(myList[0])      # Output: 10 (first element)
print(myList[-1])     # Output: 50 (last element)
print(myList[1:3])    # Output: [20, 30] (slicing)
```

Time Complexity: O(1) - constant time access using index

## Finding an Element in an Array

You have three options to find an element:

### Option 1: Direct Access (if you know the index)

Use the index value to access it directly - O(1) time complexity

```cpp
// C++
cout << myArray[3];
```

```python
# Python
print(myList[3])
```

### Option 2: Search Algorithms (if you don't know the location)

- **Linear Search** - O(n): Check each element (works on unsorted arrays)
- **Binary Search** - O(log n): Only works on sorted arrays (much faster)

**Linear Search (C++):**
```cpp
int search(int arr[], int n, int target) {
    for(int i = 0; i < n; i++) {
        if(arr[i] == target) return i;  // Found at index i
    }
    return -1;  // Not found
}
```

**Linear Search (Python):**
```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # Found at index i
    return -1  # Not found
```

**Binary Search (Python):**
```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1  # Not found
```

### Option 3: Hash-based Data Structures (for faster repeated lookups)

For multiple searches, use hash sets or hash maps - O(1) average lookup time

**Python:**
```python
my_set = {10, 20, 30, 40, 50}
if 30 in my_set:
    print("Found!")  # O(1) average time

my_dict = {"a": 1, "b": 2, "c": 3}
if "a" in my_dict:
    print("Key exists!")  # O(1) average time
```

## Array Operations and Time Complexity

### Accessing an Element
- Time: O(1) - direct access via index

### Inserting an Element

**At the end:**
- Time: O(1) - just add to the end

```python
myList.append(60)
```

**At the beginning or middle:**
- Time: O(n) - all elements need to shift

```python
myList.insert(0, 5)   # Insert at beginning
myList.insert(2, 25)  # Insert at index 2
```

### Removing an Element

**From the end:**
- Time: O(1)

```python
myList.pop()       # Remove last element
```

**From beginning or middle:**
- Time: O(n) - elements need to shift

```python
myList.pop(0)           # Remove first element
myList.remove(30)       # Remove by value
```

## Summary of Arrays

### Memory Allocation
- Memory is allocated instantly when array is created
- In C++: Fixed size, allocated at compile/runtime
- In Python: Dynamic, grows as needed

### Element Access
- Elements stored in contiguous memory locations
- Direct access via index is very efficient - O(1) constant time
- No need to search; just use the index

### Type Homogeneity
- C++: All elements must be same type (fixed, homogeneous)
- Python: Can store different types in lists (flexible, heterogeneous)

### Fixed vs Dynamic Size
- C++ arrays: Fixed size once created
- Python lists: Dynamic, can grow/shrink

### Insertion and Removal
- **At end:** O(1) - very fast
- **At beginning/middle:** O(n) - slow (elements shift)

### Best Practices
- Use arrays for fast random access
- Avoid inserting/removing from beginning/middle frequently
- For frequent insertions/deletions, consider linked lists
- Use sorted arrays + binary search for faster lookups
- Use hash-based structures for repeated lookups

### Pros and Cons

**Pros:**
- Fast access via index (O(1))
- Memory efficient
- Simple to use
- Good cache locality

**Cons:**
- Fixed size in C++
- Slow insertion/deletion in middle
- No built-in search methods
- Requires contiguous memory
