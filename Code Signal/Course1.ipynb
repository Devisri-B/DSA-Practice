# String reversal

````python
# In the range() function, a step value of -1 tells Python to count backwards instead of forwards.
# In Python, string indexing starts at 0, so the last character is always at len(string) - 1. If we started at len(string)
# range(start, stop, step), the stop value is exclusive, meaning the loop stops just before reaching it. 
# Since we want to include the character at index 0, we use -1 so the loop finishes after processing 0.
original_string = "Hello, World!"
reversed_string = ""
for i in range(len(original_string)-1, -1, -1): 
    reversed_string += original_string[i]
print(reversed_string)
````

Write a function that takes a string s, iterates through it, and collects all 0-based positions of vowels in it to a list.

For this task, consider the following letters as vowels: a, e, i, o, u (and their uppercase versions A, E, I, O, U).

Note that you should not use any Python built-in string methods to solve this task.

For example, print(solution("Hello WORLD")) should output [1, 4, 7]. Here, 'e' is a vowel, and its position in the string "Hello" is 1. 'o' is also a vowel, and its position is 4. The last vowel is O at position 7.

````python
def solution(s):
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    indices = []
    for index, ele in enumerate(s):
        if ele in vowels:
            indices.append(index)
    return indices
print(solution("Codesignal"))
````

Given a string, you need to return a new string where every letter is shifted to its right by one place in alphabetical order. The last letters z and Z should be replaced with the first ones: a and A, respectively. If the character isn't a letter, it should stay the same.

It is not allowed to use string built-in methods here.

For example, given the string "abc123XYz!", the function should return "bcd123YZa!".

````python
# To work with ASCII values in Python, you can use two helpful functions:

# ord(char) converts a character into its integer ASCII value.
# chr(value) converts an integer ASCII value back into a character.


def solution(s):
    new = []
    for ele in s:
        if ele == 'z':
            new.append('a')
        elif ele == 'Z':
            new.append('A')
        elif 'a' <= ele <= 'y' or 'A' <= ele <= 'Y':
            new.append(chr(ord(ele) + 1))
        else:
            new.append(ele)
    return "".join(new)
print(solution("Zebra-493?"))
````

# ASCII Values and Character Shifting

## ASCII Character Ranges

| Character Set | Range | Start | End |
|---|---|---|---|
| Uppercase Letters (A–Z) | 65–90 | A (65) | Z (90) |
| Lowercase Letters (a–z) | 97–122 | a (97) | z (122) |

## Shifting a Letter to the Next Character

The goal: move each letter to the next one in the alphabet, with z wrapping to a.

Think of the 26 letters as a circular loop:

```
a b c d e ... x y z → back to a
0 1 2 3 4 ... 23 24 25 → back to 0
```

## Breaking Down: 
### For lowercase
chr((ord(c) - ord('a') + 1) % 26 + ord('a'))

### For uppercase
chr((ord(c) - ord('A') + 1) % 26 + ord('A'))

### Step 1: ord(c) - ord('a')
Convert a letter into its 0-based position (0 to 25):

```
ord('a') - ord('a') = 0   → position 0
ord('b') - ord('a') = 1   → position 1
ord('z') - ord('a') = 25  → position 25
```

### Step 2: + 1
Shift to the next position:

```
'a' (pos 0) → 0 + 1 = 1
'y' (pos 24) → 24 + 1 = 25
'z' (pos 25) → 25 + 1 = 26  ← problem! position 26 doesn't exist
```

### Step 3: % 26
Wrap the position back into the 0–25 range:

```
1  % 26 = 1   → 'b'
25 % 26 = 25  → 'z'
26 % 26 = 0   → 'a' (z wraps back!)
```

### Step 4: + ord('a')
Convert the position back to an ASCII value:

```
position 0 + 97 = 97  → 'a'
position 1 + 97 = 98  → 'b'
position 25 + 97 = 122 → 'z'
```

### Step 5: chr(...)
Convert the ASCII number to a character.

## Example: c = 'z'

```
ord('z') - ord('a')  → 122 - 97 = 25
25 + 1               → 26
26 % 26              → 0       (wraps!)
0 + ord('a')         → 0 + 97 = 97
chr(97)              → 'a'
```

````python
def solution(s):
    new = []
    for ele in s:
        if 'a' <= ele <= 'z':
            new.append(chr((ord(ele) - ord('a') + 1) % 26 + ord('a')))
        elif 'A' <= ele <= 'Z':
            new.append(chr((ord(ele) - ord('A') + 1) % 26 + ord('A')))
        else:
            new.append(ele)
    return "".join(new)
print(solution("Zebra-493?"))
````

````python
# Gap between upper and lower: 'a' - 'A' = 32 — so you can convert case by adding/subtracting 32
def solution(input_string):
    output = []
    for ele in input_string:
        if ord('a') <= ord(ele) <= ord('z'):
            output.append(chr(ord(ele) - 32))
        elif ord('A') <= ord(ele) <= ord('Z'):
            output.append(chr(ord(ele) + 32))
        else:
            output.append(ele)
    return "".join(output)
print(solution("Zebra-493?"))
````

You are given a string s. Your task is to write a function that returns a string in which every pair of adjacent characters in the original string is swapped. If the string has an odd length, leave the last character as it is.

It is not allowed to use Python built-in functions like reverse() or join() in this task, you should implement the solution without using them.

For example, if you are given the string "abcdef", your function should return "badcfe". If the string is "hello", your function should return "ehllo".

````python
def solution(s):
    output = ""
    for i in range(0, len(s), 2):
        if i + 1 < len(s):
            output += s[i + 1] + s[i]
        else:
            output += s[i]
    return output
print(solution("abcdef"))
````

You are given a string, and your task is to check whether the provided string is a palindrome, without using any built-in string methods. A string is a palindrome if it reads the same forward and backward, ignoring the casing of letters ('A' and 'a' are considered the same) and ignoring non-alphanumeric characters (i.e., only letters and digits are considered, but spaces, punctuation and special characters are ignored).

Return a boolean value: True if the string is a palindrome (considering only letters and digits, and ignoring case), and False if it is not.

It is not allowed to use Python built-in functions like reverse(), reversed(), or similar in this task.

````python
def Solution(inpt_string):
    lower_case = []
    for ele in inpt_string:
        if ord('A') <= ord(ele) <= ord('Z'):
            lower_case.append(chr(ord(ele) + 32))
        else:
            lower_case.append(ele)
    left, right = 0, len(lower_case) - 1
    while left < right:
        while left < right and not ord('a') <= ord(lower_case[left]) <= ord('z') and not ord('0') <= ord(lower_case[left]) <= ord('9'):
            left += 1
        while left < right and not ord('a') <= ord(lower_case[right]) <= ord('z') and not ord('0') <= ord(lower_case[right]) <= ord('9'):
            right -= 1
        if lower_case[left] != lower_case[right]:
            return False
        else:
            left += 1
            right -= 1
    return True
print(Solution("A man, a plan, a canal: Panama"))
````

## List Comprehensions
List comprehensions are a concise way to build a list in one line instead of using a loop + append.

### Basic Syntax
```python
[expression for item in iterable]
```


### Regular Loop vs List Comprehension
```python
# Regular loop
squares = []
for x in range(5):
    squares.append(x * 2)

# Same thing as list comprehension
squares = [x * 2 for x in range(5)]
# [0, 2, 4, 6, 8]
```

### With a Condition (Filter)
```python
[expression for item in iterable if condition]

# Only even numbers
evens = [x for x in range(10) if x % 2 == 0]
# [0, 2, 4, 6, 8]
```

### With if-else (Transform)
```python
[value_if_true if condition else value_if_false for item in iterable]

# Replace negatives with 0
result = [x if x > 0 else 0 for x in [-1, 2, -3, 4]]
# [0, 2, 0, 4]
```

### Over a String
```python
chars = [c for c in "hello"]
# ['h', 'e', 'l', 'l', 'o']

upper_letters = [c for c in "hello123" if ord('a') <= ord(c) <= ord('z')]

```

### QUICK REFERENCE
| Pattern            | Syntax                          |
| ------------------ | ------------------------------- |

You are given a list of integers. Your task is to write a function find_min(nums), that returns the minimum number from the list without using Python's built-in min() function.

If the list is empty, your function should return None.

````python
def find_min(nums):
    if len(nums) == 0:
        return None
    min = nums[0]
    for num in nums:
        if num < min:
            min = num
    return min
print(find_min([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))
````

You are given an array of integers. Your task is to write a function in Python that returns the number of times the smallest element appears in the array.

Please note that built-in methods such as min() or count() should not be used in this task. Your goal is to accomplish this task by iterating over the array elements manually. Try to solve the task by doing just a single list traversal.

````python
def count_min(numbers):
    if len(numbers) == 0:
        return False
    small = numbers[0]
    count = 0
    for num in numbers:
        if num < small:
            small = num
            count = 1
        elif num == small:
            count += 1
    return count, small
        
print(count_min([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))
````

You are given a list of integers. Your task is to write a Python function to find the second-largest number among these integers. If the list has fewer than two unique numbers, return None.

You are not allowed to use any built-in Python functions or methods such as sorted(), max(), or sort(). Instead, you should implement the task using basic list operations.

````python
from typing import List, Optional

def second_max(nums: List[int]) -> Optional[int]:
    if len(nums) < 2:
         return None
    if nums[0] > nums[1]:
        first, second = nums[0], nums[1]
    elif nums[1] > nums[0]:
        first, second = nums[1], nums[0]
    else:
        first, second = nums[0], None
        
    for i in range(2, len(nums)):
        num = nums[i]
        if num > first:
            second = first
            first = num
        elif num != first:            # to handle duplicates
            if second is None or num > second:
                second = num
    return second
        
print(second_max([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))
````

# Optimizing Prime Number Checking - Using sqrt(n)

## Why Check Only Up to sqrt(n)?

Factors always come in pairs that multiply to n:

- If a number n has a factor greater than sqrt(n), it must also have a corresponding factor smaller than sqrt(n)
- Example: n = 36, sqrt(36) = 6
  - Pairs: (1, 36), (2, 18), (3, 12), (4, 9), (6, 6)
  - Every pair has one factor ≤ 6 and one ≥ 6

**Conclusion:** If no divisor exists up to sqrt(n), the number is prime. No need to check beyond.

## Concrete Example - n = 100

sqrt(100) = 10

Pairs that multiply to 100:
- (2, 50), (4, 25), (5, 20), (10, 10)

If we find that 2 divides 100, we already know 50 is a factor. We don't need to reach i=50. Checking up to 10 is enough.

## Why int(n**0.5) + 1?

- `n**0.5` - calculates the square root of n (returns a float, e.g., 3.999...)
- `int(...)` - converts to integer by truncating decimals (e.g., 3)
- `+ 1` - because range() stop is exclusive, so we add 1 to include the square root itself

### Example: n = 25

- sqrt(25) = 5.0
- `range(2, int(5.0) + 1)` = `range(2, 6)` = checks 2, 3, 4, 5 ✓
- Without +1: `range(2, int(5.0))` = `range(2, 5)` = checks 2, 3, 4 (misses 5!) ✗

**Important:** Without +1, a perfect square like 25 would be incorrectly identified as prime because its factor 5 would be skipped.

````python
def is_prime(n):
    """Function to check if n is a prime number"""
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage
print(is_prime(10)) 
print(is_prime(11))
````

````python
def next_prime(n):
    if n < 0:
        return False
    next = n + 1
    while True:
        for i in range(2, int(next ** 0.5) + 1):
            if next % i == 0:
                break
        else:
            return next
        next += 1
print(next_prime(10)) 
````

Your task is to create a Python function called get_prime_factors(n) that will return all unique prime factors of an integer n in a list. A prime factor of n is a prime number that divides n without leaving a remainder. The expected complexity is O(sqrt(n))

Note that returned prime factors should be unique and sorted in ascending order in the resulting list.

````python
def get_prime_factors(n):
    factors = []
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            factors.append(i)
            while n % i == 0:
                n = n // i
                
    if n > 1:
        factors.append(n)

    return factors
print(get_prime_factors(28))
````
