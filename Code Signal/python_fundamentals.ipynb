# Complete Guide to List Operations in Python

## What is a List?

A **list** is a collection of items that you can modify (mutable). Lists can contain any type of data: numbers, strings, other lists, mixed types, etc.

### Key Characteristics
- **Mutable**: Can change elements after creation
- **Ordered**: Elements have a specific position (index)
- **Indexed**: Access by position (0, 1, 2, ...)
- **Heterogeneous**: Can mix different data types
- **Dynamic**: Can grow or shrink

---

## Creating Lists

### Method 1: List Literal
```python
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
empty_list = []
nested = [[1, 2], [3, 4], [5, 6]]
```

### Method 2: List Constructor
```python
numbers = list()           # Empty list
from_string = list("abc")  # ['a', 'b', 'c']
from_range = list(range(5))  # [0, 1, 2, 3, 4]
```

### Method 3: List Comprehension
```python
squares = [x**2 for x in range(5)]  # [0, 1, 4, 9, 16]
evens = [x for x in range(10) if x % 2 == 0]  # [0, 2, 4, 6, 8]
```

## Accessing Elements

### By Index (Single Element)
```python
my_list = [10, 20, 30, 40, 50]

my_list[0]    # 10 (first element)
my_list[2]    # 30 (third element)
my_list[-1]   # 50 (last element)
my_list[-2]   # 40 (second to last)
```

**IndexError** if index is out of range.

### By Range (Multiple Elements - Slicing)
```python
my_list = [10, 20, 30, 40, 50]

my_list[1:3]   # [20, 30] (index 1 to 2)
my_list[:3]    # [10, 20, 30] (first 3 elements)
my_list[2:]    # [30, 40, 50] (from index 2 to end)
my_list[::2]   # [10, 30, 50] (every 2nd element)
my_list[::-1]  # [50, 40, 30, 20, 10] (reversed)
```

## Modifying Lists

### Add Elements

**append()** - Add one element to the end
```python
my_list = [1, 2, 3]
my_list.append(4)  # [1, 2, 3, 4]
```

**extend()** - Add multiple elements from another list
```python
my_list = [1, 2, 3]
my_list.extend([4, 5, 6])  # [1, 2, 3, 4, 5, 6]
```

**insert()** - Add element at specific position
```python
my_list = [1, 2, 3]
my_list.insert(1, 99)  # [1, 99, 2, 3] (insert 99 at index 1)
```

### Remove Elements

**remove()** - Remove first occurrence of value
```python
my_list = [1, 2, 3, 2, 4]
my_list.remove(2)  # [1, 3, 2, 4] (removes first 2 only)
```

**pop()** - Remove and return element at index (default: last)
```python
my_list = [1, 2, 3, 4]
last = my_list.pop()       # last = 4, my_list = [1, 2, 3]
second = my_list.pop(1)    # second = 2, my_list = [1, 3]
```

**del** - Delete element at index or slice
```python
my_list = [1, 2, 3, 4, 5]
del my_list[2]       # [1, 2, 4, 5] (delete index 2)
del my_list[1:3]     # [1, 5] (delete range)
```

**clear()** - Remove all elements
```python
my_list = [1, 2, 3]
my_list.clear()  # []
```

### Change Elements

**Direct Assignment**
```python
my_list = [1, 2, 3, 4]
my_list[0] = 100      # [100, 2, 3, 4]
my_list[1:3] = [20, 30, 40]  # [100, 20, 30, 40, 4]
```

## Common List Methods

| Method | Purpose | Example |
|--------|---------|---------|
| `append(x)` | Add x to end | `lst.append(5)` |
| `extend(lst)` | Add all elements from lst | `lst.extend([4, 5])` |
| `insert(i, x)` | Insert x at index i | `lst.insert(0, 1)` |
| `remove(x)` | Remove first x | `lst.remove(3)` |
| `pop(i)` | Remove and return element at i | `val = lst.pop()` |
| `clear()` | Remove all elements | `lst.clear()` |
| `index(x)` | Return index of first x | `idx = lst.index(3)` |
| `count(x)` | Return count of x | `cnt = lst.count(2)` |
| `sort()` | Sort in-place ascending | `lst.sort()` |
| `sort(reverse=True)` | Sort in-place descending | `lst.sort(reverse=True)` |
| `sorted(lst)` | Return new sorted list | `new_lst = sorted(lst)` |
| `reverse()` | Reverse in-place | `lst.reverse()` |
| `copy()` | Create shallow copy | `new_lst = lst.copy()` |

## Searching and Sorting

### Find Element (index)
```python
my_list = [10, 20, 30, 20, 40]

index = my_list.index(20)    # 1 (first occurrence)
# ValueError if element not found
```

### Check if Element Exists
```python
if 20 in my_list:           # True
    print("Found it")

if 100 not in my_list:      # True
    print("Not found")
```

### Count Occurrences
```python
my_list = [1, 2, 2, 3, 2, 4]
count = my_list.count(2)  # 3
```

### Sort
```python
my_list = [3, 1, 4, 1, 5]

my_list.sort()              # [1, 1, 3, 4, 5] (modifies original)
my_list.sort(reverse=True)  # [5, 4, 3, 1, 1]

new_list = sorted(my_list)  # Returns new sorted list (original unchanged)
```

### Reverse
```python
my_list = [1, 2, 3, 4, 5]

my_list.reverse()           # [5, 4, 3, 2, 1] (modifies original)
reversed_copy = my_list[::-1]  # Returns new reversed list
```

## List Comprehension

**Syntax**: `[expression for item in iterable if condition]`

### Basic Examples
```python
# Square each number
squares = [x**2 for x in range(5)]  # [0, 1, 4, 9, 16]

# Filter even numbers
evens = [x for x in range(10) if x % 2 == 0]  # [0, 2, 4, 6, 8]

# Convert strings to integers
numbers = [int(x) for x in ['1', '2', '3']]  # [1, 2, 3]

# Nested comprehension
matrix = [[i*j for j in range(3)] for i in range(3)]
# [[0, 0, 0], [0, 1, 2], [0, 2, 4]]
```

### Comparison: Loop vs Comprehension

**Using loop:**
```python
result = []
for x in range(5):
    if x % 2 == 0:
        result.append(x**2)
```

**Using comprehension (more Pythonic):**
```python
result = [x**2 for x in range(5) if x % 2 == 0]
```

## List vs String vs Tuple

| Property | List | String | Tuple |
|----------|------|--------|-------|
| Mutable | YES | NO | NO |

## Time Complexity of List Operations

| Operation | Complexity | Notes |
|-----------|-----------|-------|
| Access by index | O(1) | Direct access |
| Append | O(1) amortized | Usually fast |
| Insert at beginning | O(n) | Requires shift |
| Insert at middle | O(n) | Requires shift |
| Remove (by value) | O(n) | Must search first |
| Pop last | O(1) | Fast |
| Pop first | O(n) | Requires shift |
| Search (index/in) | O(n) | Linear search |
| Sort | O(n log n) | Python's Timsort |
| Copy | O(n) | Creates new list |

## Important Gotchas

### Gotcha 1: Assignment vs Copy
```python
list1 = [1, 2, 3]
list2 = list1        # Points to same list (alias)
list2.append(4)
print(list1)         # [1, 2, 3, 4] - CHANGED!

# Correct way to copy:
list2 = list1.copy()        # Shallow copy
list2 = list1[:]            # Shallow copy
from copy import deepcopy
list2 = deepcopy(list1)     # Deep copy (for nested lists)
```

### Gotcha 2: Default Mutable Arguments
```python
def bad_function(items=[]):  # DANGER!
    items.append(1)
    return items

print(bad_function())  # [1]
print(bad_function())  # [1, 1] - Same list reused!

# Correct:
def good_function(items=None):
    if items is None:
        items = []
    items.append(1)
    return items
```

### Gotcha 3: List Comprehension Scope
```python
# x is accessible outside (unlike some languages)
[x for x in range(5)]
print(x)  # 4 - x still exists!
```

## Best Practices

1. Use `append()` for adding single items
2. Use `extend()` for adding multiple items (not `+` in loops)
3. Use list comprehension for creating new lists
4. Use `in` operator to check existence (not `index()` with try/except)
5. Use `enumerate()` when you need both index and value
6. Use slicing `[:]` instead of `list()` for shallow copy
7. Be careful with mutable default arguments
8. Don't modify a list while iterating over it
9. Use `sorted()` when you need a new list; `sort()` for in-place
10. Remember: lists are mutable, strings are not
